```python
from bs4 import BeautifulSoup
import requests
import time
from bin.scrape import Make_scraper, makes_dict_save
```

```python
url = 'https://www.otomoto.pl/sitemap-categories-make-osobowe.xml'
response = requests.get(url)
soup = BeautifulSoup(response.content, features="xml")
loc = soup.find_all('loc')

makes_lst = []  #list of links to different makes
for i in loc:
    makes_lst.append(i.text)
print(len(makes_lst))
```

```python
makes_dict = {}
for i in makes_lst:
    
    time.sleep(0.15)
    make = Make_scraper(i)

    num_text = make.check_len()
    pos1 = num_text.find(' - ')
    pos2 = num_text.find(' ogł')
    if -1 not in [pos1, pos2]:
        num = num_text[pos1+3:pos2]
    else:
        print(i, 'err')
        continue
    if int(num.replace(' ', '')) > 175000:
        continue
    else:
        makes_dict[i] = num
```

```python
missing_lst = []
for link in makes_lst:
    if link not in makes_dict.keys():
        missing_lst.append(link)
```

```python
for elem in range(len(missing_lst)):
    # missing_lst[elem] = missing_lst[elem].replace('+', '-')
    if missing_lst[elem] == 'https://www.otomoto.pl/osobowe/%C5%A0koda/':
        missing_lst[elem] = 'https://www.otomoto.pl/osobowe/skoda/'
```
