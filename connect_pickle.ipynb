```python
import pandas as pd
from os import listdir
from os.path import isfile, join
```

```python
mypath = './data/'
```

```python
onlyfiles = [mypath + f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
onlyfiles
```

```python
onlyfiles = onlyfiles[:-2]
```

```python
onlyfiles
```

```python
data = pd.read_pickle(onlyfiles[0])
print(onlyfiles[0])
for file in onlyfiles[1:]:
    print(file)
    data = pd.concat([data, pd.read_pickle(file)])
```

```python
data.shape[0]
```

```python
data = data.drop_duplicates()
data.shape[0]
```
